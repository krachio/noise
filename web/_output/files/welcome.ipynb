```python
import piplite
await piplite.install("faust-dsl")
await piplite.install("krach")
from krach._web_audio import connect_web
import krach.dsp as krs
kr = connect_web(bpm=120, master=0.7)
print("krach ready! kr and krs are in scope.")
```

```python
# Create a voice and play a pattern
bass = kr.node("bass", "oscillator", gain=0.3)
bass @ kr.hit() * 4
print("bass playing!")
```

```python
# Sequence
bass @ kr.seq("A2", "D3", None, "E2").over(2)
```

```python
# Swing
bass @ (kr.hit() * 8).swing(0.67)
```

```python
# Route to an effect
verb = kr.node("verb", "oscillator", gain=0.2)
bass >> verb
```

```python
# Control
bass["cutoff"] = 1200
```

```python
# Stop
bass @ None
```
